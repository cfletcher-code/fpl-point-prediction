```python
# Select specific columns from 'gw_df' DataFrame and create 'cleaned_gw_df'
cleaned_gw_df = gw_df[['GW', 'season', 'team', 'kickoff_time', 'opponent_team', 'was_home', 'name', 'position', 'minutes', 'bonus', 'bps', 'total_points', 'clean_sheets']]

# Merge 'cleaned_gw_df' with 'teams_df' on 'season' and 'team' columns
cleaned_gw_df = cleaned_gw_df.merge(teams_df[['season', 'id', 'name', 'strength']], left_on=['season', 'team'], right_on=['season', 'name'], how='left')

# Rename columns in 'cleaned_gw_df': 'strength' to 'team_strength', 'name_y' to 'team_name', and 'id' to 'team_id'
cleaned_gw_df = cleaned_gw_df.rename(columns={"strength": "team_strength", "name_y": "team_name", "id": "team_id"})

# Merge 'cleaned_gw_df' again with 'teams_df' on 'season' and 'opponent_team' columns
cleaned_gw_df = cleaned_gw_df.merge(teams_df[['season', 'id', 'name', 'strength']], left_on=['season', 'opponent_team'], right_on=['season', 'id'], how='left')

# Rename columns in 'cleaned_gw_df': 'strength' to 'opponent_strength', 'name_x' to 'name', 'name' to 'opponent_team_name', and 'id' to 'opponent_team_id'
cleaned_gw_df = cleaned_gw_df.rename(columns={"strength": "opponent_strength", "name_x": "name", "name": "opponent_team_name", "id": "opponent_team_id"})

# Select specific columns in 'cleaned_gw_df'
cleaned_gw_df = cleaned_gw_df[['GW', 'season', 'kickoff_time', 'team', 'team_id', 'opponent_team_name', 'opponent_team_id', 'team_strength', 'opponent_strength', 'was_home', 'name', 'position', 'minutes', 'bonus', 'bps', 'total_points', 'clean_sheets']]

# Rename the 'clean_sheets' column to 'output_clean_sheets'
cleaned_gw_df = cleaned_gw_df.rename({'clean_sheets': 'output_clean_sheets'}, axis=1)

cleaned_gw_df = cleaned_gw_df[~cleaned_gw_df['position'].isna()]

cleaned_gw_df.columns
```

```python
# Select specific columns from 'fixtures_df' DataFrame and create reduced 'fixtures_red_df'
fixtures_red_df = fixtures_df[['event', 'kickoff_time', 'team_a', 'team_h']]

# Rename columns in 'fixtures_red_df' to prepare 'fixtures_away_df'
fixtures_away_df = fixtures_red_df.rename(columns={'event': 'GW', 'kickoff_time': 'kickoff_time', 'team_a': 'team', 'team_h': 'opponent_team_id'})
fixtures_away_df['is_home'] = False  # Add a new column 'is_home' and set it to False

# Rename columns in 'fixtures_red_df' to prepare 'fixtures_home_df'
fixtures_home_df = fixtures_red_df.rename(columns={'event': 'GW', 'kickoff_time': 'kickoff_time', 'team_h': 'team', 'team_a': 'opponent_team_id'})
fixtures_home_df['is_home'] = True  # Add a new column 'is_home' and set it to True

# Concatenate 'fixtures_home_df' and 'fixtures_away_df' to create 'all_season_fixtures_df'
all_season_fixtures_df = pd.concat([fixtures_home_df, fixtures_away_df])

# Filter out rows where 'kickoff_time' is NaN
all_season_fixtures_df = all_season_fixtures_df[~all_season_fixtures_df['kickoff_time'].isna()]

# Convert the 'GW' column to a string data type
all_season_fixtures_df['GW'] = all_season_fixtures_df['GW'].astype('str')

# Select specific columns from 'players_raw' DataFrame and create 'current_players'
current_players = players_raw[['first_name', 'second_name', 'team', 'now_cost', 'element_type']]

# Create a 'name' column by combining 'first_name' and 'second_name'
current_players['name'] = current_players['first_name'] + ' ' + current_players['second_name']

# Map 'element_type' values to position abbreviations (e.g., 1:'GK', 2:'DEF', etc.)
current_players['element_type'] = current_players['element_type'].map({1: 'GK', 2: 'DEF', 3: 'MID', 4: 'FWD'})

# Drop the 'first_name' and 'second_name' columns
current_players = current_players.drop(['first_name', 'second_name'], axis=1)

# Add a 'season' column with the value '2023-24'
current_players['season'] = '2023-24'

# Merge 'all_season_fixtures_df' and 'current_players' on the 'team' column
this_season_gw = all_season_fixtures_df.merge(current_players, on='team')

# Rename columns: 'element_type' to 'position', 'is_home' to 'was_home', and 'opponent_team_id' to 'opponent_team'
this_season_gw = this_season_gw.rename({'element_type': 'position', 'is_home': 'was_home', 'opponent_team_id': 'opponent_team'}, axis=1)

# Select specific columns in 'this_season_gw' DataFrame
this_season_gw = this_season_gw[['GW', 'season', 'team', 'opponent_team', 'was_home', 'name', 'position', 'kickoff_time']]

# Rename columns in 'teams_df' DataFrame and merge with 'this_season_gw'
teams = teams_df.rename({'name': 'team_name'}, axis=1)
this_season_gw = this_season_gw.merge(teams[['season', 'id', 'team_name', 'strength']], left_on=['season', 'team'], right_on=['season', 'id'], how='left')

# Rename columns in 'this_season_gw' DataFrame and merge with 'teams_df' again for 'opponent_team'
this_season_gw = this_season_gw.rename(columns={"strength": "team_strength", "id": "team_id"})
this_season_gw['team'] = this_season_gw['team_name']
this_season_gw = this_season_gw.merge(teams[['season', 'id', 'team_name', 'strength']], left_on=['season', 'opponent_team'], right_on=['season', 'id'], how='left')

# Rename columns and select specific columns in 'this_season_gw' DataFrame
this_season_gw = this_season_gw.rename(columns={"strength": "opponent_strength", "id": "opponent_team_id"})
this_season_gw['opponent_team_name'] = this_season_gw['team_name_y']

# Remove rows where 'GW' is NaN, convert 'GW' to int, and select specific columns
this_season_gw = this_season_gw[~this_season_gw['GW'].isna()]
this_season_gw['GW'] = this_season_gw['GW'].astype(float).astype(int)
this_season_gw = this_season_gw[['GW', 'season', 'kickoff_time', 'team', 'team_id', 'opponent_team_name', 'opponent_team_id', 'team_strength', 'opponent_strength', 'was_home', 'name', 'position']]

# Convert 'kickoff_time' column to datetime data type
this_season_gw['kickoff_time'] = pd.to_datetime(this_season_gw['kickoff_time'])
```
